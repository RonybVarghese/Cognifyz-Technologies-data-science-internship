```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Load the dataset to inspect its structure
file_path = (r'C:\Users\Rony\Downloads\Dataset .csv')
data = pd.read_csv(file_path)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for the plots
sns.set(style="whitegrid")

# Plot: Distribution of Ratings
plt.figure(figsize=(10, 6))
sns.histplot(data['Aggregate rating'], bins=20, kde=True, color='blue', alpha=0.7)
plt.title("Distribution of Aggregate Ratings", fontsize=16)
plt.xlabel("Aggregate Rating", fontsize=14)
plt.ylabel("Frequency", fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()

```

# 2. Bar plot of rating counts

```python

plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='Rating text', order=data['Rating text'].value_counts().index, palette="viridis")
plt.title("Count of Ratings by Text", fontsize=16)
plt.xlabel("Rating Text", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.show()
```

# 3. Average ratings by city (Top 10 cities with most restaurants)

```python

city_avg_ratings = data.groupby('City')['Aggregate rating'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(14, 7))
sns.barplot(x=city_avg_ratings.values, y=city_avg_ratings.index, palette='viridis')
plt.title('Average Ratings by City (Top 10)', fontsize=16)
plt.xlabel('Average Rating', fontsize=12)
plt.ylabel('City', fontsize=12)
plt.show()
```

# 4. Average ratings by cuisine (Top 10 cuisines)

```python

data['Cuisines'] = data['Cuisines'].fillna('Unknown')
cuisine_avg_ratings = data.groupby('Cuisines')['Aggregate rating'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(14, 7))
sns.barplot(x=cuisine_avg_ratings.values, y=cuisine_avg_ratings.index, palette='coolwarm')
plt.title('Average Ratings by Cuisine (Top 10)', fontsize=16)
plt.xlabel('Average Rating', fontsize=12)
plt.ylabel('Cuisine', fontsize=12)
plt.show()
```

# 5. Relationship between Votes and Ratings

```python

plt.figure(figsize=(10, 6))
sns.scatterplot(x=data['Votes'], y=data['Aggregate rating'], alpha=0.6, color='darkorange')
plt.title('Relationship between Votes and Aggregate Rating', fontsize=16)
plt.xlabel('Votes', fontsize=12)
plt.ylabel('Aggregate Rating', fontsize=12)
plt.show()
```

## Summary of Visualizations:
### Distribution of Ratings: The histogram and KDE plot reveal that most restaurants are rated between 3.0 and 4.5, with relatively few low or perfect ratings.

### Average Ratings by City: Among the top 10 cities with the most restaurants, some cities have notably higher average ratings, indicating potential regional preferences or quality differences.

### Average Ratings by Cuisine: The analysis of top-rated cuisines shows that certain cuisines consistently perform better in terms of aggregate ratings.

### Relationship between Votes and Ratings: A positive trend suggests that higher-rated restaurants tend to receive more votes, although some variability exists.
